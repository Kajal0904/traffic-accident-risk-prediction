```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("SpotifyFeatures.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
```

```python
features = [
    'danceability', 'energy', 'loudness',
    'speechiness', 'acousticness',
    'instrumentalness', 'liveness',
    'valence', 'tempo'
]

X = df[features].dropna()
df = df.loc[X.index]
```

```python
X.hist(figsize=(12,10), bins=20)
plt.suptitle("Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(X.sample(500))
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=X['energy'], y=X['loudness'])
plt.title("Energy vs Loudness")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=X['danceability'], y=X['valence'])
plt.title("Danceability vs Happiness (Valence)")
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df['PCA1'] = X_pca[:,0]
df['PCA2'] = X_pca[:,1]

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set2')
plt.title("Clusters Visualization (PCA)")
plt.show()
```

```python
cluster_summary = df.groupby('Cluster')[features].mean()
print(cluster_summary)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(cluster_summary.drop('Type', axis=1), annot=True, cmap='coolwarm')
plt.title("Cluster Feature Comparison")
plt.show()
```

```python
for feature in features:
    plt.figure(figsize=(2,2))
    sns.boxplot(x='Cluster', y=feature, data=df)
    plt.title(f"{feature} Distribution per Cluster")
    plt.show()
```

```python
sns.countplot(x='Cluster', data=df)
plt.title("Number of Songs in Each Cluster")
plt.show()
```

```python
for i in range(5):
    print(f"\nCluster {i} sample songs:")
    print(df[df['Cluster']==i]['track_name'].head(5))
```

```python
def describe_cluster(row):
    if row['energy'] > 0.7:
        return "High Energy"
    elif row['acousticness'] > 0.7:
        return "Acoustic"
    elif row['instrumentalness'] > 0.5:
        return "Instrumental"
    else:
        return "Mixed"

cluster_summary['Type'] = cluster_summary.apply(describe_cluster, axis=1)
print(cluster_summary[['Type']])
```
